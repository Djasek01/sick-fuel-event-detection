## Definiranje značajki i preprocesora

U ovom koraku definiraju se ulazne značajke za modeliranje te se postavlja **preprocessing pipeline** koji automatski obrađuje numeričke i kategoričke varijable.

- Numeričke značajke: `fuel_lvl`, `speed`, `gps_latitude`, `gps_longitude`
- Kategorička značajka: `contact_value`

Za numeričke značajke primjenjuje se imputacija medijanom i standardizacija, dok se za kategoričku značajku koristi imputacija najčešćom vrijednošću i One-Hot enkodiranje. Svi ovi koraci objedinjeni su u `ColumnTransformer` objektu `preprocessor`, koji se kasnije koristi unutar modelskih pipelineova (Naivni Bayes, Random Forest, XGBoost).

Ovaj dio izravno pokriva: 
- obradu nedostajućih vrijednosti
- skaliranje numeričkih varijabli
- enkodiranje kategoričke varijable `contact_value` (One-Hot Encoding)


```python
# Definiranje značajki
numerical_features = ['fuel_lvl', 'speed', 'gps_latitude', 'gps_longitude', 'fuellvlprev', 'deltafuel']
categorical_features = ['contact_value']

# Pipeline za numeričke značajke: Imputacija medijanom + Standardizacija
num_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

# Pipeline za kategoričke značajke: Imputacija najčešćom vrijednošću + One-Hot Encoding
cat_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

# Spajanje u ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_transformer, numerical_features),
        ('cat', cat_transformer, categorical_features)
    ]
)

# Prikaz preprocesora
print("Preprocessing pipeline je definiran.")
display(preprocessor)
```

## Definiranje ulaznih značajki i ciljne varijable

U ovom koraku jasno se razdvajaju ulazne značajke (X) i ciljna varijabla (y) za klasifikacije.

## Podjela na trening i test skup

Podaci se zatim dijele na trening (70 %) i test skup (30 %) uz korištenje parametra `stratify=y`.  
Stratifikacija osigurava da raspodjela klasa u trening i test skupu ostane slična kao u originalnom skupu, što je posebno važno zbog **debalansa klasa** u ciljnoj varijabli.

```python
# Definiranje značajki (X) i ciljne varijable (y)
X = merged.drop(columns=['withdrawl_flag', 'time', 'obj_id'])
y = merged['withdrawl_flag']

# Podjela na trening i test skup (70/30)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42, stratify=y)

print(f"Veličina skupa za treniranje: {X_train.shape[0]} redaka")
print(f"Veličina skupa za testiranje: {X_test.shape[0]} redaka")
```

## Uzorkovanje podataka za modeliranje

Zbog vrlo velikog skupa podataka, u ovom koraku se nasumično izdvaja uzorak od 200 000 zapisa (`merged_sample_200`) kako bi treniranje modela bilo računski izvedivo.
Na tom uzorku ponovno se definiraju ulazne značajke (`X_s2`) i ciljna varijabla (`y_s2`), pri čemu se, kao i ranije, iz značajki uklanjaju stupci `withdrawl_flag`, `time` i `obj_id`.

Podaci iz uzorka zatim se dijele na trening i test skup (70/30) uz stratifikaciju po `y_s2`, kako bi se zadržala ista struktura klasa i u manjem skupu.  
Time i dalje poštuješ zahtjev zadatka vezan uz **debalans klasa**, ali u okvirima skupa koji je dovoljno mali za brže eksperimente s modelima.

## Naivni Bayes – prvi model predviđanja

Na uzorku definira se i trenira Naivni Bayes model koristeći već ranije definirani `preprocessor`:

- `preprocessor` automatski:
  - imputira nedostajuće vrijednosti,
  - skalira numeričke značajke,
  - One-Hot enkodira kategoričku varijablu `contact_value`.
- `GaussianNB` je klasifikator koji modelira ciljnu varijablu `withdrawl_flag`.

Nakon treniranja modela, na test skupu se izračunavaju sljedeće metrike:
- točnost (Accuracy),
- ponderirana preciznost (Precision),
- ponderirani odziv (Recall),
- ponderirani F1-score,
- višeklasni ROC AUC (strategija `ovr`).

```python
# Uzorkovanje 200,000 zapisa
merged_sample_200 = merged.sample(n=200000, random_state=42)
X_s2 = merged_sample_200.drop(columns=['withdrawl_flag', 'time', 'obj_id'])
y_s2 = merged_sample_200['withdrawl_flag']

# Podjela na trening i test (70/30)
X_train_2, X_test_2, y_train_2, y_test_2 = train_test_split(X_s2, y_s2, test_size=0.30, random_state=42, stratify=y_s2)

# Definiranje i treniranje modela
nb_pipeline_200 = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', GaussianNB())
])

nb_pipeline_200.fit(X_train_2, y_train_2)

# Predviđanje i evaluacija
y_pred_200 = nb_pipeline_200.predict(X_test_2)
y_proba_200 = nb_pipeline_200.predict_proba(X_test_2)

# Izračun metrika
accuracy = accuracy_score(y_test_2, y_pred_200)
precision = precision_score(y_test_2, y_pred_200, average='weighted', zero_division=0)
recall = recall_score(y_test_2, y_pred_200, average='weighted')
f1 = f1_score(y_test_2, y_pred_200, average='weighted')
roc_auc = roc_auc_score(y_test_2, y_proba_200, multi_class='ovr')

# Prikaz rezultata u traženom formatu
print("--- Rezultati Naivnog Bayesa ---")
print(f"Accuracy:  {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}")
print(f"F1-score:  {f1:.4f}")
print(f"ROC AUC:   {roc_auc:.4f}")
print("\nClassification Report:")
print(classification_report(y_test_2, y_pred_200))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Definiranje oznaka klasa na temelju y_test_2
labels = sorted(y_test_2.unique())

# Izračun matrice zabune
cm_nb_200 = confusion_matrix(y_test_2, y_pred_200, labels=labels)

# Vizualizacija
plt.figure(figsize=(8, 6))
sns.heatmap(cm_nb_200, annot=True, fmt='d', cmap='Blues',
            xticklabels=labels, yticklabels=labels)
plt.title('Matrica zabune: Naivni Bayes (200k uzorak)')
plt.xlabel('Predviđena klasa')
plt.ylabel('Stvarna klasa')
plt.show()
```

Za Random Forest računa se i prikazuje **matrica zabune** (`cm_rf_200`) koristeći iste oznake klasa (`labels`) kao kod Naivnog Bayesa.
Vizualizacija putem `seaborn.heatmap` (zeleno obojena skala) omogućuje pregled gdje Random Forest postiže bolje ili lošije razvrstavanje u odnosu na "Naivni Bayes", posebno u kontekstu debalansa klasa.

```python
from sklearn.ensemble import RandomForestClassifier

# Definiranje i treniranje Random Forest modela na 200k uzorku
rf_pipeline_200 = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(n_estimators=100, random_state=50, n_jobs=-1))
])

rf_pipeline_200.fit(X_train_2, y_train_2)

# Predviđanje i evaluacija
y_pred_rf_200 = rf_pipeline_200.predict(X_test_2)
y_proba_rf_200 = rf_pipeline_200.predict_proba(X_test_2)

# Izračun metrika
accuracy_rf = accuracy_score(y_test_2, y_pred_rf_200)
precision_rf = precision_score(y_test_2, y_pred_rf_200, average='weighted', zero_division=0)
recall_rf = recall_score(y_test_2, y_pred_rf_200, average='weighted')
f1_rf = f1_score(y_test_2, y_pred_rf_200, average='weighted')
roc_auc_rf = roc_auc_score(y_test_2, y_proba_rf_200, multi_class='ovr')

# Prikaz rezultata
print("--- Rezultati Random Foresta ---")
print(f"Accuracy:  {accuracy_rf:.4f}")
print(f"Precision: {precision_rf:.4f}")
print(f"Recall:    {recall_rf:.4f}")
print(f"F1-score:  {f1_rf:.4f}")
print(f"ROC AUC:   {roc_auc_rf:.4f}")
print("\nClassification Report:")
print(classification_report(y_test_2, y_pred_rf_200))

# Matrica zabune
plt.figure(figsize=(8, 6))
cm_rf_200 = confusion_matrix(y_test_2, y_pred_rf_200, labels=labels)
sns.heatmap(cm_rf_200, annot=True, fmt='d', cmap='Greens',
            xticklabels=labels, yticklabels=labels)
plt.title('Matrica zabune: Random Forest (200k uzorak)')
plt.xlabel('Predviđena klasa')
plt.ylabel('Stvarna klasa')
plt.show()
```

## XGBoost kao napredni model predviđanja

U ovom dijelu uvodi se **XGBoost** kao treći, napredni model predviđanja na istom uzorku od 200 000 zapisa.  
Budući da XGBoost očekuje da oznake klasa budu kodirane cjelobrojnim vrijednostima koje počinju od 0, ciljana varijabla `y_train_2` i `y_test_2` najprije se transformiraju pomoću `LabelEncoder`a (`y_train_xgb_200`, `y_test_xgb_200`).

Na taj način XGBoost dobiva na ulazu već pripremljene, standardizirane i enkodirane značajke, jednako kao i Naivni Bayes te Random Forest.

Nakon treniranja, izračunavaju se iste metrike kao za ostale modele:

- ponderirani F1-score

## Vizualizacija matrice zabune za XGBoost

Za XGBoost se računa i prikazuje **matrica zabune** (`cm_xgb_200`), pri čemu se:
- koristi ista struktura (stvarna vs. predviđena klasa),
- oznake na osima (`xticklabels` i `yticklabels`) postavljaju na originalne klase pomoću `le_200.classes_`.

Vizualizacija putem `seaborn.heatmap` (narančasta skala) omogućuje pregled raspodjele točnih i pogrešnih predikcija XGBoosta te usporedbu s matricama zabune za Naivni Bayes i Random Forest.  


```python
from xgboost import XGBClassifier
from sklearn.preprocessing import LabelEncoder

# XGBoost zahtijeva da klase počinju od 0 (npr. 0, 1, 2)
le_200 = LabelEncoder()
y_train_xgb_200 = le_200.fit_transform(y_train_2)
y_test_xgb_200 = le_200.transform(y_test_2)

# Kreiranje XGBoost modela u pipelineu
xgb_pipeline_200 = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', XGBClassifier(n_estimators=100, random_state=50, use_label_encoder=False, eval_metric='mlogloss'))
])

# Treniranje
xgb_pipeline_200.fit(X_train_2, y_train_xgb_200)

# Predviđanje i vjerojatnosti
y_pred_xgb_200 = xgb_pipeline_200.predict(X_test_2)
y_proba_xgb_200 = xgb_pipeline_200.predict_proba(X_test_2)

# Izračun metrika
accuracy_xgb = accuracy_score(y_test_xgb_200, y_pred_xgb_200)
precision_xgb = precision_score(y_test_xgb_200, y_pred_xgb_200, average='weighted', zero_division=0)
recall_xgb = recall_score(y_test_xgb_200, y_pred_xgb_200, average='weighted')
f1_xgb = f1_score(y_test_xgb_200, y_pred_xgb_200, average='weighted')
roc_auc_xgb = roc_auc_score(y_test_xgb_200, y_proba_xgb_200, multi_class='ovr')

# Prikaz rezultata
print("--- Rezultati XGBoosta ---")
print(f"Accuracy:  {accuracy_xgb:.4f}")
print(f"Precision: {precision_xgb:.4f}")
print(f"Recall:    {recall_xgb:.4f}")
print(f"F1-score:  {f1_xgb:.4f}")
print(f"ROC AUC:   {roc_auc_xgb:.4f}")
print("\nClassification Report:")
print(classification_report(y_test_xgb_200, y_pred_xgb_200, target_names=[str(c) for c in le_200.classes_]))

# Matrica zabune
plt.figure(figsize=(8, 6))
cm_xgb_200 = confusion_matrix(y_test_xgb_200, y_pred_xgb_200)
sns.heatmap(cm_xgb_200, annot=True, fmt='d', cmap='Oranges',
            xticklabels=le_200.classes_, yticklabels=le_200.classes_)
plt.title('Matrica zabune: XGBoost (200k uzorak)')
plt.xlabel('Predviđena klasa')
plt.ylabel('Stvarna klasa')
plt.show()
```

Matrice zabune dodatno pokazuju koliko se razlikuju ove klase: dijagonalni elementi predstavljaju točne pogodke po klasi, dok izvan-dijagonalni elementi otkrivaju gdje modeli najviše griješe (npr. kad sumnjive događaje (-1) pogrešno klasificiraju kao normalne (0)).

## Usporedna analiza performansi modela

U ovom završnom koraku izrađuje se **sažetak performansi sva tri modela** (Naivni Bayes, Random Forest, XGBoost) u jednoj tablici i vizualizaciji.  
Najprije se numeričke vrijednosti metrika (Accuracy, Precision, Recall, F1-score, ROC AUC) za svaki model spremaju u `DataFrame` `df_comparison`, a zatim se pomoću `melt` transformacije pretvaraju u dugi format (`df_melted`) pogodan za grupirani stupčasti graf.

Pomoću `seaborn.barplot` crta se graf na kojem:
- X-os predstavlja modele (Naivni Bayes, Random Forest, XGBoost),
- Y-os predstavlja vrijednosti metrika,
- boje (hue) razlikuju pojedine metrike (Accuracy, Precision, Recall, F1-score, ROC AUC).

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Osiguravanje da su podaci u ispravnom formatu za vizualizaciju
comparison_data = {
    'Model': ['Naivni Bayes', 'Random Forest', 'XGBoost'],
    'Accuracy': [accuracy, accuracy_rf, accuracy_xgb],
    'Precision': [precision, precision_rf, precision_xgb],
    'Recall': [recall, recall_rf, recall_xgb],
    'F1-score': [f1, f1_rf, f1_xgb],
    'ROC AUC': [roc_auc, roc_auc_rf, roc_auc_xgb]
}
df_comparison = pd.DataFrame(comparison_data)
df_melted = df_comparison.melt(id_vars='Model', var_name='Metrika', value_name='Vrijednost')

# Vizualizacija s modelima na X-osi prema vašem kodu
plt.figure(figsize=(14, 8))
sns.barplot(data=df_melted, x='Model', y='Vrijednost', hue='Metrika', palette='magma')

plt.title('Usporedba performansi po modelima (200k uzorak)')
plt.ylim(0, 1.1)
plt.ylabel('Vrijednost metrike')
plt.xlabel('Model')
plt.legend(title='Metrike', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Dodavanje vrijednosti iznad stupaca
for p in plt.gca().patches:
    if p.get_height() > 0:
        plt.gca().annotate(f'{p.get_height():.3f}',
                       (p.get_x() + p.get_width() / 2., p.get_height()),
                       ha = 'center', va = 'center',
                       xytext = (0, 9),
                       textcoords = 'offset points',
                       fontsize=8,
                       rotation=0)

plt.tight_layout()
plt.show()
```
